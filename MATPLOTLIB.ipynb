Matplotlit - useful for making plots and graphs

```python
#importing the matplotlib library
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,10, 100)
y = np.sin(x)
z = np.cos(x)
```

Plotting the data

```python
#sin wave
plt.plot(x,y)
plt.show()
```

```python
#cos wave
plt.plot(x,z)
plt.show()
```

```python
#Adding tittle, x-axis & y-axis label to our line graph
plt.plot(x,y)
plt.xlabel('Angle')
plt.ylabel('Sine value')
plt.title('Sine wave')
plt.show()
```

```python
plt.plot(x,z)
plt.xlabel('Angle')
plt.ylabel('Cosine value')
plt.title('Cosine wave')
plt.show()
```

```python
#Ploting a parabola
x = np.linspace(-10,10, 100)
y = x**2
plt.plot(x,y)
plt.xlabel('X values')
plt.ylabel('Y values')
plt.title('Parabola')
plt.show()
```

```python
plt.plot(x, y, 'g+')
plt.show()
```

```python
x = np.linspace(-5,5, 100)
plt.plot(x, np.sin(x), 'r +')
plt.plot(x, np.cos(x), 'g')
plt.xlabel('X values')
plt.ylabel('Y values')
plt.title('Cosine and Sine graph')
plt.show()
```
